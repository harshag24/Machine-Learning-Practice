**For each unique customer, order frequency => number of unique orders/years of engagement + years of engagement**

```python
df_unique_dates = df.groupby('Email').apply(lambda g: g['Date'].unique())
df_unique_dates
```

```python
dict_uniq_date = df_unique_dates.to_dict()
df2 = pd.DataFrame(dict_uniq_date.keys(), columns={'Email'})
```

```python
df2['Dates'] = dict_uniq_date.values()
df2
```

```python
from datetime import datetime

date_format = '%m/%d/%Y'
today = datetime.strptime('10/1/2021', date_format)
c=0
# df2['Email'][c],
for i in df2['Dates']:
    arr = i.tolist()
    all_dates = []
    if c<13326:
        for j in arr:
            if len(arr)>1:
                if j == arr[len(arr)-1]:
                    all_dates.append(j)
                    all_dates.sort(key = lambda date: datetime.strptime(date, date_format))
                    min_date_order = datetime.strptime(all_dates[0], date_format)
                    max_date_order = datetime.strptime(all_dates[len(arr)-1], date_format)
                    diff = today - min_date_order 
                    engament = max_date_order - min_date_order

                    print('Order Frequencyy with >1 order -',df2['Email'][c],' - ', df_3[c]/(diff.days/365)) # Can use engament instead of diff.days if years of engament is required
                    print('Years of engamenet -',df2['Email'][c],' - ', engament.days/365)
                else:
                    all_dates.append(j)
            
            else:
                date_order = datetime.strptime(j, date_format)
                diff = today - date_order
                print('Order Frequency with only 1 order -',df2['Email'][c],' - ',df_3[c]/(diff.days/365))
                print('Years of engamenet -',df2['Email'][c],' - ', diff.days/365)
                
    c = c+1   
```

**Avg order value for each customer= total amount of purchase done / total number of orders placed**

```python
df_sum = df.groupby('Email').apply(lambda g: pd.Series(g['Net Sales']).sum())
df_sum
```

```python
dict_sum = df_sum.to_dict()
df4 = pd.DataFrame(dict_sum.keys(), columns={'Email'})
df4['Sum'] = dict_sum.values()
df4
```

```python
df4['Sum'] = df4['Sum'].str.replace('$','')
df4['Sum'] = df4['Sum'].str.replace('-','0')
df4['Sum'] = df4['Sum'].str.replace('(','')
df4['Sum'] = df4['Sum'].str.replace(')',' ')
df4['Sum'] = df4['Sum'].str.replace(',','')

df4
```

```python
orders = []
sum_orders = []

for i in df4['Sum']:
    x = i.split()
    orders.append(len(x))
    int_list = [int(j) for j in x]
    sum_orders.append(sum(int_list))
```

```python
c=0
for i in df4['Email']:
    print('Avg Order Value for ',i, ' is ', sum_orders[c]/orders[c])
    c = c+1
```
